```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate random points
x1 = np.random.normal(0, 1, 5000)  # Gaussian distribution
x2 = np.random.uniform(-3, 3, 5000)  # Uniform distribution

# Create a figure with two subplots
plt.figure(figsize=(12, 5))

# Gaussian distribution subplot
plt.subplot(121)
plt.hist(x1, bins=50, edgecolor='black', color='blue', alpha=0.7)
plt.title('Gaussian Distribution Frequency')
plt.xlabel('X')
plt.ylabel('Frequency')
plt.grid(True)

# Uniform distribution subplot
plt.subplot(122)
plt.hist(x2, bins=50, edgecolor='black', color='red', alpha=0.7)
plt.title('Uniform Distribution Frequency')
plt.xlabel('X')
plt.ylabel('Frequency')
plt.grid(True)

plt.tight_layout()
plt.show()

# Print distribution statistics
print("Gaussian Distribution:")
print(f"Mean: {np.mean(x1):.4f}")
print(f"Standard Deviation: {np.std(x1):.4f}")

print("\nUniform Distribution:")
print(f"Mean: {np.mean(x2):.4f}")
print(f"Standard Deviation: {np.std(x2):.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate random points from normal distribution
num_points = 5000
x = np.random.normal(0, 1, num_points)  # mean=0, std=1

# Create histogram
plt.figure(figsize=(10, 6))
plt.hist(x, bins=50, edgecolor='black')
plt.title('Frequency Distribution of Normal Random Points')
plt.xlabel('X Values')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Display distribution statistics
print(f"Mean: {np.mean(x):.4f}")
print(f"Standard Deviation: {np.std(x):.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 10,000 random numbers between 1 and 100
random_numbers = np.random.uniform(1, 100, 10000)

# Plot the distribution
plt.figure(figsize=(10, 6))
plt.hist(random_numbers, bins=50, color="blue", alpha=0.7, edgecolor='black')

# Add labels and title
plt.title("Distribution of 10,000 Random Numbers (1 to 100)", fontsize=16)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Frequency", fontsize=14)

# Show the plot
plt.show()

# Optional: Print distribution statistics
print(f"Mean: {np.mean(random_numbers):.2f}")
print(f"Standard Deviation: {np.std(random_numbers):.2f}")
print(f"Min: {np.min(random_numbers):.2f}")
print(f"Max: {np.max(random_numbers):.2f}")
```
