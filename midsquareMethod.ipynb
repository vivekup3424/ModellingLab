```python
import numpy as np

def mid_square_method(seed, num_points=100):
    """
    Generate random numbers using Mid-Square method
    
    Parameters:
    - seed: Initial 4-digit number
    - num_points: Number of random numbers to generate
    
    Returns:
    - List of generated random numbers
    """
    random_numbers = []
    current = seed
    
    # Create a log file to track details
    with open('mid_square_log.txt', 'w') as log_file:
        for n in range(num_points):
            # Square the current number
            squared = current ** 2
            
            # Convert to string to easily extract middle digits
            squared_str = str(squared).zfill(8)  # Pad to 8 digits
            
            # Extract middle 4 digits
            current = int(squared_str[2:6])
            
            # Normalize to 0-9999 range
            current = current % 10000
            
            # Normalized random number
            random_value = current / 10000
            
            # Log details
            log_entry = (f"Iteration {n+1}: "
                         f"Seed = {current}, "
                         f"Squared = {squared}, "
                         f"Random Value = {random_value}, "
                         f"2*n = {2*(n+1)}")
            
            log_file.write(log_entry + '\n')
            print(log_entry)
            
            random_numbers.append(random_value)
    
    return random_numbers

# Set a seed for reproducibility
np.random.seed(42)

# Generate random points
seed_value = 5690737
random_points = mid_square_method(seed_value)

# Visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(range(len(random_points)), random_points, alpha=0.7, color='blue')
plt.title('Random Points Generated by Mid-Square Method')
plt.xlabel('Point Index')
plt.ylabel('Random Value')
plt.grid(True)
plt.show()

# Additional information
print("\nMid-Square Method Algorithm Steps:")
print("1. Start with a 4-digit seed number")
print("2. Square the seed number")
print("3. Extract the middle 4 digits of the squared number")
print("4. Use these digits as the new seed")
print("5. Repeat the process to generate more random numbers")
print(f"\nUsed seed: {seed_value}")
print(f"Generated {len(random_points)} random points between 0 and 1")
print("\nDetailed log has been saved to 'mid_square_log.txt'")
```
