```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate random points
x1 = np.random.normal(0, 1, 100)    # Gaussian distribution
x2 = np.random.uniform(0, 1, 100)   # Uniform distribution
x3 = np.random.exponential(1/2, 100)  # Exponential distribution with lamda = 2

# Create a figure with three subplots
plt.figure(figsize=(15, 5))

# Gaussian distribution subplot
plt.subplot(131)
plt.hist(x1, bins=20, edgecolor='black', color='blue', alpha=0.7)
plt.title('Normal Distribution')
plt.xlabel('X')
plt.ylabel('Frequency')
plt.grid(True)

# Uniform distribution subplot
plt.subplot(132)
plt.hist(x2, bins=20, edgecolor='black', color='red', alpha=0.7)
plt.title('Uniform Distribution')
plt.xlabel('X')
plt.ylabel('Frequency')
plt.grid(True)

# Exponential distribution subplot
plt.subplot(133)
plt.hist(x3, bins=20, edgecolor='black', color='green', alpha=0.7)
plt.title('Exponential Distribution ( = 2)')
plt.xlabel('X')
plt.ylabel('Frequency')
plt.grid(True)

plt.tight_layout()
plt.show()

# Print distribution statistics
print("Gaussian Distribution:")
print(f"Mean: {np.mean(x1):.4f}")
print(f"Standard Deviation: {np.std(x1):.4f}")

print("\nUniform Distribution:")
print(f"Mean: {np.mean(x2):.4f}")
print(f"Standard Deviation: {np.std(x2):.4f}")

print("\nExponential Distribution:")
print(f"Mean: {np.mean(x3):.4f}")
print(f"Standard Deviation: {np.std(x3):.4f}")
```
