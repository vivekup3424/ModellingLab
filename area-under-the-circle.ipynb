```python
import numpy as np
import matplotlib.pyplot as plt

def is_inside_circle(x, y):
    # Check if point (x,y) is inside or on the circle x²+y²=4
    return x**2 + y**2 <= 4

def monte_carlo_circle_area(num_points=int(1e5)):
    # Bounding square is from -2 to 2 in both x and y
    x = np.random.uniform(-2, 2, num_points)
    y = np.random.uniform(-2, 2, num_points)
    
    # Points inside the circle
    points_inside = is_inside_circle(x, y)
    
    # Calculate area
    total_area = 4 * 4  # area of bounding square
    circle_area = total_area * np.sum(points_inside) / num_points
    
    return circle_area, points_inside, x, y

# Perform integration
result, points_inside, x, y = monte_carlo_circle_area()

# Visualization
plt.figure(figsize=(10, 6))
plt.scatter(x[points_inside], y[points_inside], c='blue', alpha=0.1, label='Points inside circle')
plt.scatter(x[~points_inside], y[~points_inside], c='red', alpha=0.1, label='Points outside circle')

# Draw circle
circle = plt.Circle((0, 0), 2, fill=False, color='green', label='Circle x²+y²=4')
plt.gca().add_artist(circle)

plt.title(f'Monte Carlo Integration: Area = {result:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.legend()
plt.show()

print(f"Estimated Area: {result:.2f}")

# Analytical solution
print(f"Analytical Area (π*r²): {np.pi * 4:.2f}")
```
